```python
from uszipcode import SearchEngine
```

```python
search = SearchEngine(simple_zipcode=False)
```

```python
result = search.by_coordinates(38.683304, -121.164457, radius=10, returns=1)
```

```python
print(result[0].zipcode)
```

```python
import csv
```

```python
with open('latlongs.csv') as csv_file:
    csv_reader = csv.reader(csv_file, delimiter=',')
    line_count = 0
    for row in csv_reader:
        result = search.by_coordinates(float(row[0]), float(row[1]), radius=20, returns=1)
        
        print(row[0] + ", " + row[1] + ", " + result[0].zipcode)
        line_count += 1
        
    print(f'Processed {line_count} lines.')
```
